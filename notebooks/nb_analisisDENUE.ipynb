```python
import pandas as pd
import geopandas as gpd

pd.set_option('display.max_columns', None)
```

```python
denue = gpd.read_file('../datos/01_procesos/unidades_economicas_2025.gpkg')
denue.head(1)
```

```python
denue['rama'].unique()
```

```python
# Clasificación de ramas económicas relacionadas con el CDIT Vallejo-i
# Agrupadas en 6 categorías temáticas

ramas_cdit = {
    'Tecnología, informática y telecomunicaciones': [
        'Servicios de diseño de sistemas de cómputo y servicios relacionadosT ',
        'Servicios de investigación científica y desarrollo',
        'Edición de software',
        'Provisión de infraestructura de servicios de cómputo, procesamiento de datos, hospedaje de páginas de internet y otros servicios relacionados',
        'Operadores de servicios de telecomunicaciones alámbricas e inalámbricas',
        'Otros servicios de telecomunicaciones',
        'Operadores de servicios de telecomunicaciones vía satéliteT ',
        'Bibliotecas, archivos, portales de búsqueda en la red y otros servicios de suministro de información',
        'Comercio al por menor de mobiliario, equipo y accesorios de cómputo, teléfonos y otros aparatos de comunicación',
        'Comercio al por mayor de mobiliario y equipo de cómputo y de oficina, y de otra maquinaria y equipo de uso general',
    ],
    'Diseño y economía creativa': [
        'Diseño especializado',
        'Servicios de publicidad y actividades relacionadas',
        'Artistas, escritores y técnicos independientes',
        'Compañías y grupos de espectáculos artísticos y culturales',
        'Promotores de espectáculos artísticos, culturales, deportivos y similares',
        'Industria fílmica y del videoT ',
        'Industria del sonido',
        'Edición de periódicos, revistas, libros, directorios y otros materiales',
        'Museos, sitios históricos, zoológicos y similares',
    ],
    'Manufactura tecnológica e industria avanzada': [
        'Fabricación de computadoras y equipo periférico',
        'Fabricación de componentes electrónicos',
        'Fabricación de equipo de comunicación',
        'Fabricación de equipo de audio y de video',
        'Fabricación de instrumentos de medición, control, navegación, y equipo médico electrónico',
        'Fabricación de maquinaria y equipo para el comercio y los servicios',
        'Fabricación de equipo de generación y distribución de energía eléctrica',
        'Fabricación de aparatos eléctricos de uso doméstico',
        'Fabricación y reproducción de medios magnéticos y ópticos',
        'Fabricación de maquinaria y equipo para la industria metalmecánica',
        'Impresión e industrias conexas',
    ],
    'Servicios profesionales, científicos y técnicos': [
        'Servicios de arquitectura, ingeniería y actividades relacionadas',
        'Otros servicios profesionales, científicos y técnicos',
        'Servicios de consultoría administrativa, científica y técnica',
        'Servicios de administración de negocios',
        'Dirección y administración de grupos empresariales o corporativos',
        'Servicios de contabilidad, auditoría y servicios relacionados',
        'Servicios legales',
    ],
    'Educación y capacitación': [
        'Escuelas de educación superior',
        'Escuelas de educación técnica superior',
        'Escuelas comerciales, de computación y de capacitación para ejecutivos',
        'Escuelas de oficios',
        'Otros servicios educativos',
        'Servicios de capacitación para el trabajo para personas desempleadas, subempleadas o con discapacidad',
    ],
    'Apoyo a negocios y emprendimiento': [
        'Servicios integrales de apoyo a los negocios en instalaciones',
        'Otros servicios de apoyo a los negocios',
        'Servicios de empleo',
        'Servicios de apoyo secretarial, fotocopiado, cobranza, investigación crediticia y similares',
    ],
}

# Mapeo inverso: rama -> categoría
rama_a_categoria = {
    rama: categoria
    for categoria, ramas in ramas_cdit.items()
    for rama in ramas
}

print(f"Total de categorías: {len(ramas_cdit)}")
print(f"Total de ramas clasificadas: {len(rama_a_categoria)}")
```

```python
# Agregar columna de categoría CDIT y filtrar unidades relacionadas
denue['categoria_cdit'] = denue['rama'].map(rama_a_categoria)

denue_cdit = denue[denue['categoria_cdit'].notna()].copy()

print(f"Total unidades económicas en la zona: {len(denue):,}")
print(f"Unidades relacionadas con el CDIT:    {len(denue_cdit):,}")
print(f"Proporción: {len(denue_cdit)/len(denue)*100:.1f}%")
print()
print("Distribución por categoría:")
conteo = denue_cdit.groupby('categoria_cdit').size().sort_values(ascending=False).rename('unidades')
print(conteo.to_string())
```

```python
# Detalle por rama dentro de cada categoría
detalle = (
    denue_cdit
    .groupby(['categoria_cdit', 'rama'])
    .size()
    .reset_index(name='unidades')
    .sort_values(['categoria_cdit', 'unidades'], ascending=[True, False])
)
detalle
```
