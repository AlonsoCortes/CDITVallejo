```python
import pandas as pd
import geopandas as gpd
pd.set_option('display.max_columns', None)
```

```python
denue = gpd.read_file('../docs/datos/unidades_economicas_cdit_2025.geojson')
denue.head(1)
```

```python
denue['categoria_cdit'].value_counts()
```

```python
# edu = denue[denue['categoria_cdit'].str.contains('edu', case=False, na=False)]
edu = denue[denue['categoria_cdit'] == 'Educación y capacitación']

print(edu.shape)
edu.head(1)
```

```python
edu['subrama'].value_counts()
```

```python
categorias = ['Escuelas de educación primaria', 
              'Escuelas de educación secundaria general',
              'Escuelas de educación secundaria técnica',
              'Escuelas de educación media técnica terminal',
              'Escuelas de educación media superior', 
              'Escuelas de educación superior',
              'Escuelas de educación técnica superior', 
              'Escuelas que combinan diversos niveles de educación']

edu_1 = edu[edu['subrama'].isin(categorias)].copy()

nivel_map = {
    'Escuelas de educación primaria':                        'Educación básica',
    'Escuelas de educación secundaria general':              'Educación básica',
    'Escuelas de educación secundaria técnica':              'Educación básica',
    'Escuelas de educación media superior':                  'Educación media',
    'Escuelas de educación media técnica terminal':          'Educación media',
    'Escuelas de educación superior':                        'Educación superior',
    'Escuelas de educación técnica superior':                'Educación superior',
    'Escuelas que combinan diversos niveles de educación':   'Multinivel',
}

edu_1['nivel_educativo'] = edu_1['subrama'].map(nivel_map)

edu_1['nivel_educativo'].value_counts()
```

```python
edu.to_crs(epsg=4326).to_file('../docs/datos/ue_educativas_cdit_2025.geojson')
```
